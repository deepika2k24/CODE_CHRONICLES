###### 3D Contour plot

```python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z,75, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

```python
# changing the view angle
ax.view_init(60, 35)
fig
```

###### Geographic Data with Basemap

```python
"""One common type of visualization in data science is that of geographic data. Matplotlib’s
main tool for this type of visualization is the Basemap toolkit, which is one of
several Matplotlib toolkits that live under the mpl_toolkits namespace."""
%pip install basemap
```

```python
from mpl_toolkits.basemap import Basemap
```

```python
# Create a figure with a specific size (8x8 inches)
plt.figure(figsize=(8, 8))

# Initialize a Basemap projection with an orthographic view
# 'ortho' stands for Orthographic projection, which gives a 3D-like view of the Earth
# 'resolution=None' means no boundaries like coastlines will be drawn for faster rendering
# 'lat_0' and 'lon_0' set the central latitude and longitude of the projection
# This projection is centered at latitude 50 and longitude -100
m = Basemap(projection='ortho', resolution=None, 
            lat_0=50, lon_0=-100)

# Add the "Blue Marble" background image to the map
# 'bluemarble' is a realistic satellite imagery of Earth
# 'scale=0.5' reduces the resolution of the imagery for better performance
m.bluemarble(scale=0.5)

# Display the map
plt.show()
```

```python
# Create a figure with a specific size (8x8 inches)
fig = plt.figure(figsize=(8, 8))

# Initialize a Basemap projection centered at a latitude and longitude
# 'lcc' stands for Lambert Conformal Conic projection
# 'resolution=None' avoids drawing coastlines for faster rendering
# 'width' and 'height' define the size of the map in meters
# 'lat_0' and 'lon_0' set the central latitude and longitude of the projection
m = Basemap(projection='lcc', resolution=None,
            width=8E6, height=8E6,
            lat_0=45, lon_0=-100)

# Add topography to the map with ETOPO data
# 'scale=0.5' determines the scale of the ETOPO topography data
# 'alpha=0.5' sets the transparency level of the topography overlay
m.etopo(scale=0.5, alpha=0.5)

# Map the coordinates of Seattle, USA (longitude, latitude) to the projection's x, y coordinates
# Seattle's coordinates: longitude -122.3, latitude 47.6
x, y = m(-122.3, 47.6)

# Plot a marker at Seattle's location
# 'ok' indicates a black circular marker
# 'markersize=5' sets the marker size to 5
plt.plot(x, y, 'ok', markersize=5)

# Add a text label for Seattle at the specified coordinates
# 'fontsize=12' sets the font size of the text
plt.text(x, y, ' Seattle', fontsize=12)

# Display the map
plt.show()
```
