```python
#prg1A
import matplotlib.pyplot as plt
import numpy as np

# Data
fuel_types = ['Petrol', 'Diesel', 'Electric', 'Hybrid', 'CNG']
avg_prices = [25000, 27000, 35000, 30000, 20000]
colors = ['SkyBlue', 'Orange', 'Green', 'Red', 'Purple']

# Create the plot
plt.figure(figsize=(8, 5))
bars = plt.bar(fuel_types, avg_prices, color=colors)

# Rotate x-axis labels if any fuel type label is long
rotation_angle = 0 if max([len(f) for f in fuel_types]) < 5 else 45
plt.xticks(rotation=rotation_angle)

# Add gridlines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Annotate each bar with price label
for bar, price in zip(bars, avg_prices):
    plt.text(bar.get_x() + bar.get_width() / 2,
             price + 500,
             f"${price}",
             ha='center',
             fontsize=10,
             fontweight='bold')

# Labels and title
plt.title("Average Car Prices by Fuel Type", fontsize=14, fontweight='bold')
plt.xlabel("Fuel Type", fontsize=12)
plt.ylabel("Average Price ($)", fontsize=12)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
#prg5
import warnings
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats
warnings.simplefilter(action='ignore', category=FutureWarning)
np.random.seed(42)
x = np.random.rand(100) * 100
y = 2.5 * x + np.random.normal(0, 25, 100)
data_dataframe = pd.DataFrame({'x': x, 'y': y})
sns.set_theme(style="whitegrid", font_scale=1.2)
plt.figure(figsize=(10, 6))

ax = sns.regplot(
 data=data_dataframe,
 x='x',
 y='y',
 ci=95,
 scatter_kws={
 's': 60,
 'alpha': 0.7,
 'edgecolor': 'k',
 'linewidths': 0.5, # fixed here
 'marker': 'o'
 },
 line_kws={
 'color': 'crimson',
 'linewidth': 2.5,
 'alpha': 0.9
 }
)

ax.set_title('Enhanced Regression Plot', fontsize=16, weight='bold', pad=15)
ax.set_xlabel('x', fontsize=14)
ax.set_ylabel('y', fontsize=14)
sns.rugplot(data=data_dataframe, x='x', ax=ax, color='navy', height=0.03)
sns.rugplot(data=data_dataframe, x='y', ax=ax, color='navy', height=0.03)
slope, intercept, r_value, p_value, std_err = stats.linregress(data_dataframe['x'],
data_dataframe['y'])

reg_eq = f'y = {slope:.2f}x + {intercept:.2f}\n$R^2$ = {r_value**2:.2f}'
ax.text(
 0.05, 0.95, reg_eq,
 transform=ax.transAxes,
 fontsize=12,
 verticalalignment='top',
 bbox=dict(boxstyle='round,pad=0.4', facecolor='white', alpha=0.5)
)
plt.tight_layout()
plt.show()
```

```python
#prg6
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import HoverTool, Label, ColumnDataSource
from bokeh.layouts import column
output_notebook()

data = {
 'x_values': [1, 2, 3, 4, 5, 6],
 'y_values_1': [10, 20, 15, 25, 20, 30],
 'y_values_2': [5, 15, 20, 10, 25, 20],
}
source = ColumnDataSource(data)
p = figure(title="Enhanced Line Graph", x_axis_label='X-Axis', y_axis_label='Y-Axis',
 width=800, height=400)
line1 = p.line(x='x_values', y='y_values_1', source=source, line_color='blue',
 line_width=2, legend_label='Series 1', line_dash='solid')
p.circle(x='x_values', y='y_values_1', source=source, size=8, color='blue', alpha=0.6)
line2 = p.line(x='x_values', y='y_values_2', source=source, line_color='red',
 line_width=2, legend_label='Series 2', line_dash='dotted')
p.circle(x='x_values', y='y_values_2', source=source, size=8, color='red', alpha=0.6)
annotations = [
 ('Point A', 3, 15),
 ('Point B', 5, 25),
]

for label, x, y in annotations:
 p.add_layout(Label(x=x, y=y, text=label, text_font_size='10pt',
 text_color='black'))
hover = HoverTool()

hover.tooltips = [
 ("X", "$x"),
 ("Value Series 1", "@y_values_1"),
 ("Value Series 2", "@y_values_2"),
]

p.add_tools(hover)
p.legend.location = 'top_left'
p.legend.title = 'Legend'
p.legend.title_text_font_style = 'bold'
p.legend.label_text_font_size = '10pt'
p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = 'gray'
p.ygrid.grid_line_alpha = 0.5
show(p)
```

```python
#prg7
from bokeh.io import output_notebook, show
output_notebook()
from bokeh.layouts import column, row
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.plotting import figure
from bokeh.palettes import Category10

x = [1, 2, 3, 4, 5]
y_line = [2, 4, 6, 8, 10]
y_scatter = [1, 5, 3, 7, 2]
categories = ['A', 'B', 'C', 'D', 'E']
bar_values = [5, 7, 2, 6, 4]

colors = Category10[5]
line_source = ColumnDataSource(data=dict(x=x, y=y_line))
scatter_source = ColumnDataSource(data=dict(x=x, y=y_scatter))
bar_source = ColumnDataSource(data=dict(x=categories, top=bar_values, color=colors))

line_plot = figure(title="Line Plot", x_axis_label="X", y_axis_label="Y", width=400, height=300)
line_plot.line('x', 'y', source=line_source, line_width=2, color="navy", legend_label="Line")
line_plot.circle('x', 'y', source=line_source, size=8, color="navy")
hover_line = HoverTool(tooltips=[("X", "@x"), ("Y", "@y")])
line_plot.add_tools(hover_line)
line_plot.legend.location = "top_left"

scatter_plot = figure(title="Scatter Plot", x_axis_label="x", y_axis_label="y", width=400, height=300)
scatter_plot.scatter('x', 'y', source=scatter_source, size=10, color="green", marker="diamond",
legend_label="scatter")
hover_scatter = HoverTool(tooltips=[("X", "@x"), ("Y", "@y")])
scatter_plot.add_tools(hover_scatter)
scatter_plot.legend.location = "top_left"

bar_plot = figure(x_range=categories, title="Bar Plot", y_axis_label="value", width=800, height=300)
bar_plot.vbar(x='x', top='top', width=0.6, source=bar_source, fill_color='color')
hover_bar = HoverTool(tooltips=[("Category", "@x"), ("Value", "@top")])
bar_plot.add_tools(hover_bar)
layout = column(row(line_plot, scatter_plot), bar_plot)


show(layout)
```

```python
#prg8
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from PIL import Image

image_path = r"/content/Screenshot 2025-08-17 112701.png"

original_img = Image.open(image_path)
gray_img = original_img.convert('L')
gray_array = np.array(gray_img)

x = np.arange(0, gray_array.shape[1])
y = np.arange(0, gray_array.shape[0])
X, Y = np.meshgrid(x, y)
Z = gray_array

plt.figure(figsize=(6, 6))
plt.imshow(original_img)
plt.title('Original Color Image')
plt.axis('off')
plt.show()

plt.figure(figsize=(6, 6))
plt.imshow(gray_array, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()

fig1 = plt.figure(figsize=(10, 7))
ax1 = fig1.add_subplot(111, projection='3d')
ax1.plot_surface(X, Y, Z, cmap='gray', edgecolor='none')
ax1.set_title('3D Surface Plot of Grayscale Image')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Grayscale Intensity')
plt.show()

sine_x = np.linspace(0, gray_array.shape[1], 500)
sine_y = np.linspace(0, gray_array.shape[0], 500)
sine_X, sine_Y = np.meshgrid(sine_x, sine_y)
sine_Z = 50 * np.sin(0.1 * sine_X)

fig2 = plt.figure(figsize=(10, 7))
ax2 = fig2.add_subplot(111, projection='3d')
ax2.plot_surface(sine_X, sine_Y, sine_Z + np.mean(Z), cmap='viridis', alpha=0.6)
ax2.set_title('3D Sine Wave Projection')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Amplitude')
plt.show()
```

```python
#prg9
import plotly.graph_objects as go
import pandas as pd
import numpy as np

np.random.seed(42)
dates = pd.date_range('2023-01-01', periods=100, freq='D')
data = np.cumsum(np.random.randn(100))
fig = go.Figure()
fig.add_trace(go.Scatter(
    x=dates,
    y=data,
    mode='lines+markers',
    name='Sample Data',
    line=dict(color='royalblue', width=2),
    marker=dict(size=5, color='red', symbol='circle')
))
fig.update_layout(
    title='Optimized Time Series Plot',
    title_x=0.5,
    xaxis=dict(
        title='Date',
        showgrid=True,
        zeroline=False,
        tickangle=45,
        tickformat='%Y-%m-%d'
    ),
    yaxis=dict(
        title='Value',
        showgrid=True,
        zeroline=False
    ),
    plot_bgcolor='white',
    hovermode='closest',
    autosize=True,
    margin=dict(l=50, r=50, t=50, b=50)
)
fig.add_annotation(
    x=dates[50],
    y=data[50],
    text="Midpoint of the data",
    showarrow=True,
    arrowhead=2,
    ax=-50,
    ay=-50
)

fig.show()
```

```python
#PRG6
from bokeh.io import output_notebook, show
output_notebook()
from bokeh.plotting import figure, show
from bokeh.models import Label, HoverTool, Span, ColumnDataSource
from bokeh.layouts import layout
import pandas as pd
import numpy as np

# Generate sample time series data
dates = pd.date_range(start="2024-01-01", periods=50)
values1 = np.cumsum(np.random.randn(50)) + 20
values2 = np.cumsum(np.random.randn(50)) + 10

# Create data sources
source1 = ColumnDataSource(data={'x': dates, 'y': values1})
source2 = ColumnDataSource(data={'x': dates, 'y': values2})

# Initialize the figure
p = figure(title="Enhanced Time Series Line Plot with Annotations",
           x_axis_type="datetime",
           width=900,
           height=450,
           sizing_mode='stretch_width',
           tools="pan,wheel_zoom,box_zoom,reset,save",
           toolbar_location="above")
# Plot the two lines
line1 = p.line('x', 'y', source=source1, line_width=3, color='blue', legend_label='Metric A')
line2 = p.line('x', 'y', source=source2, line_width=3, color='green', legend_label='Metric B', line_dash="dashed")

# Add a hover tool
hover = HoverTool(
    tooltips=[
        ("Date", "@x{%F}"),
        ("Value", "@y{0.2f}")
    ],
    formatters={"@x": 'datetime'},
    mode='vline'
)

p.add_tools(hover)

# Add an annotation at the maximum value of the first line
max_index = np.argmax(values1)
max_date = dates[max_index]
max_value = values1[max_index]

# Add Label
label = Label(x=max_date, y=max_value, text=f"Peak: {max_value:.2f}",
              text_font_size="10pt", background_fill_color='white',
              text_color='black', text_baseline="bottom")

p.add_layout(label)

vline = Span(location=max_date.timestamp() * 1000,
             dimension='height', line_color='red',
             line_dash='dotted', line_width=2)
p.add_layout(vline)

# Customize axes
p.xaxis.axis_label = "Date"
p.yaxis.axis_label = "Metric Value"
p.xaxis.axis_label_text_font_size = "12pt"
p.yaxis.axis_label_text_font_size = "12pt"
p.xaxis.major_label_orientation = 0.8

# Customize Legend
p.legend.title = "Metrics"
p.legend.label_text_font_size = "10pt"
p.legend.title_text_font_style = "bold"
p.legend.location = "top_left"
p.legend.click_policy = "hide"  # Click to hide/show lines

# Display the plot
show(layout([p], sizing_mode='stretch_width'))

```
